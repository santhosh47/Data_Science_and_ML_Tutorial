##**What are Outliers?**
In terms of statistics, Outliers can be defined as,

‚ÄúAn Outlier is that observation which is significantly different from all other observations.‚Äù

From this definition, we can conclude that an outlier is something that is an odd-one-out or the one that is different from the crowd. Some statisticians formally define outliers as **‚ÄòObservations having a different underlying behavior than the rest of the observations‚Äô.**

Alternatively, outliers are those observations that are significantly different from other observations.

##**When are outliers dangerous?**
Outliers are not always dangerous for our problem statement. In fact, outliers sometimes can be helpful indicators.

They represent errors in the measurement, bad data collection(not careful while data collection), or simply show those variables that are not considered while collecting the data. Many data analysts are directly tempted to delete outliers. However, this is sometimes the wrong choice for our predictive analysis. One cannot recognize outliers while collecting the data for the problem statement; you won‚Äôt know what data points are outliers until you begin analyzing the data. Since some of the statistical tests are sensitive to the outliers and therefore, the ability to detect them and treat them accordingly is an important part of data analytics.

##**Which statistics are affected by the outliers?**
Let‚Äôs discuss one by one for each statistic,

üëâ **Mean:** It is the only measure of central tendency that is always affected by an outlier since it is calculated as the sum of the observed values and then divide by the total number of observations. Since in the expression of mean, the total sum is included, and due to outliers, there are some abnormal values i.e. Outliers will affect this sum.

For Example, the Let‚Äôs outlier is having a bigger positive value than the other values that will make the sum large enough so that the mean will also be slightly larger while if the outlier has a very small value, then the mean will also become a bit smaller. Hence the presence of outliers in our dataset can largely affect the mean.

üëâ **Standard deviation (SD)**: It is calculated with the help of every observation in the data set. It is a sensitive measure because it will be influenced by outliers since standard deviation is calculated by taking the difference of sample case from the mean, outliers will affect Standard deviation.

üëâ **Median:** The median is defined as the middle value in a particular distribution. It is the data point at which half of the observations are above, and half of the observations are below wrt that point. It is not affected by outliers, therefore the median is preferred as a measure of central tendency when a distribution has extreme observations.

üëâ **Inter-Quartile Range (IQR):** The IQR is the difference between the 75th and 25th percentile. The IQR is more resistant to outliers. The IQR by definition only covers the middle 50% of the data, so outliers are well outside this range and the presence of a small number of outliers is not likely to change this significantly. If you add an outlier, the IQR will change to another set of data points that are probably not that dissimilar to the previous ones (in most datasets), hence it is ‚Äúresistant‚Äù to change. This is especially the case of a large dataset.

Now if you add some crazy extreme data point at the end, the 75th and 25th percentile doesn‚Äôt change much, because extreme outliers or no, 75% of the data still lies below roughly the same amount.

üëâ**Range:** Most affected by the outliers since it is the difference b/w the max and min value present in the dataset.

##**When to drop or keep outliers?**
I believe that the dropping outlier is always a harsh step and should be taken only in extreme conditions when we‚Äôre very sure that the outlier is due to a measurement error, which we generally do not know while doing analysis.

Sometimes outliers indicate a mistake in data collection. Other times, though, they can influence a data set, so it‚Äôs important to keep them to better understand the dataset in the big picture.

Below are some examples that give you a clear idea about when you should and shouldn‚Äôt drop outliers.

##**Drop an outlier if:**

üëâ **You know that it‚Äôs completely wrong**

For example, if you have a really good sense of how range our data should fall in, like people‚Äôs ages, which we discussed above in scenario-1, you can safely drop values outside of that range.

üëâ **You have a lot of data in hand**

When you have a lot of data in your hands, then your sample won‚Äôt be hurt by dropping a questionable outlier.

üëâ **You have an option to going back**

You can go back and recollect and verify the questionable observations.

##**Don‚Äôt drop an outlier if:**

üëâ **Your results are critical**

 When your results are critical, then even minor changes will matter a lot.

For example, You can feel better about dropping outliers of the dataset in which there are people‚Äôs favorite TV shows, but not about the temperatures at which airplane seals fail.

üëâ **There are a lot of outliers**

By definition, Outliers are rare.

For example, Let‚Äôs 25% of your data be outliers, then it means that something is interesting going on with your data that you need to look further into. You can relate this with scenario-2 which we discussed in the above section.

##**How to treat outliers?**
üëâ **Trimming:** It excludes the outlier values from our analysis. By applying this technique our data becomes thin when there are more outliers present in the dataset. Its main advantage is its fastest nature.

üëâ**Capping:** In this technique, we cap our outliers data and make the limit i.e, above a particular value or less than that value, all the values will be considered as outliers, and the number of outliers in the dataset gives that capping number.

For Example, if you‚Äôre working on the income feature, you might find that people above a certain income level behave in the same way as those with a lower income. In this case, you can cap the income value at a level that keeps that intact and accordingly treat the outliers.

üëâ**Treat outliers as a missing value:** By assuming outliers as the missing observations, treat them accordingly i.e, same as those of missing values.

[You can refer to the missing value article](https://www.analyticsvidhya.com/blog/2021/04/how-to-handle-missing-values-of-categorical-variables/) 

üëâ **Discretization:** In this technique, by making the groups we include the outliers in a particular group and force them to behave in the same manner as those of other points in that group. This technique is also known as Binning.

[You can learn more about discretization](https://www.analyticsvidhya.com/blog/2021/05/complete-guide-on-encode-numerical-features-in-machine-learning/) 

##**How to detect outliers?**
üëâ **For Normal distributions:** Use empirical relations of Normal distribution.

‚Äì The data points which fall below mean-3*(sigma) or above mean+3*(sigma) are outliers.

where mean and sigma are the average value and standard deviation of a particular column.

image.png

üëâ **For Skewed distributions:** Use Inter-Quartile Range (IQR) proximity rule.

‚Äì The data points which fall below Q1 ‚Äì 1.5 IQR or above Q3 + 1.5 IQR are outliers.

where Q1 and Q3 are the 25th and 75th percentile of the dataset respectively, and IQR represents the inter-quartile range and given by Q3 ‚Äì Q1.

üëâ**For Other distributions:** Use percentile-based approach.

For Example, Data points that are far from 99% percentile and less than 1 percentile are considered an outlier.

üëâ **Z-score treatment :**

Assumption‚Äì The features are normally or approximately normally distributed.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Step-2: Read and Load the Dataset

df = pd.read_csv('placement.csv')
df.sample(5)
```

```python
#Step-3: Plot the Distribution plots for the features

import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(df['cgpa'])
plt.subplot(1,2,2)
sns.distplot(df['placement_exam_marks'])
plt.show()
```

image.png

```python
#Step-4: Finding the Boundary Values

print("Highest allowed",df['cgpa'].mean() + 3*df['cgpa'].std())
print("Lowest allowed",df['cgpa'].mean() - 3*df['cgpa'].std())
```

Output:

Highest allowed 8.808933625397177

Lowest allowed 5.113546374602842

```python
#Step-5: Finding the Outliers

df[(df['cgpa'] > 8.80) | (df['cgpa'] < 5.11)]
```

```python
#Step-6: Trimming of Outliers

new_df = df[(df['cgpa'] < 8.80) & (df['cgpa'] > 5.11)]
new_df
```

```python
#Step-7: Capping on Outliers

upper_limit = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
#Step-8: Now, apply the Capping

df['cgpa'] = np.where(
    df['cgpa']>upper_limit,
    upper_limit,
    np.where(
        df['cgpa']<lower_limit,
        lower_limit,
        df['cgpa']
    )
)
```

```python
#Step-9: Now see the statistics using ‚ÄúDescribe‚Äù Function

df['cgpa'].describe()
```
