# Basics of logistic regression

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set()

#Apply a fix to the statsmodels library
from scipy import stats
stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq, df)
```

## Load the data

```python
raw_data = pd.read_csv(r"C:\Users\santh\Downloads\DS_70_Admittance_Dataset.csv")
data = raw_data.copy()
data['Admitted'] = raw_data['Admitted'].map({'Yes': 1, 'No': 0})
data
```

## Declare the dependent and the independent variables

```python
y = data['Admitted']
x1 = data['SAT']
```

## Regression

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
```

## Summary

```python
# Get the regression summary
results_log.summary()
```

The logic model for the SAT Admittance Regression
we created is
log of pi divided by
one minus pi equals minus 69.91 plus 0.042 times SAT.

Pi here refers to the probability
of an event occurring.
While one minus pi to the probability
of the event not occurring.
The fraction of these two is a very popular concept
odds as we explained earlier.



Alright, so from the regression summary
it seems that our model is basically
log of the odds equals minus 69.91 plus 0.042 times SAT.

Let's go through the process of interpreting this result.

First, I'll take two SAT scores, SAT one and SAT two.
If we plug them in, we would get two equations,
one for each of them.

Log of odds two equals minus 69.91 plus 0.042 times SAT two.
Log of odds one equals minus 69.91 plus 0.042 times SAT one.

Now let's create their difference.

Log of odds two minus log of odds one equals
0.042 times SAT two minus SAT one in brackets.

Looks much better, doesn't it?
Now the difference between the logs
is the log of the quotient of their arguments.

Therefore, log of odds two divided by odds one
equals 0.042 times SAT two minus SAT one in brackets.

Okay, now let's take a difference in the SAT score of one.
So SAT two minus SAT one equals one.

Log of odds two divided by odds one equals
0.042 times one or 0.042.

To solve this equation, we must take the exponential
of the two sides as the exponential of log of X
equals X by definition.

In this way, we will simplify the expression further.
In the end, we get odds two divided by odds one equals
E to the power of 0.042 or 1.042.

### Plot a logistic regression curve

```python
# Creating a logit regression (we will discuss this in another notebook)
reg_log = sm.Logit(y,x)
# Fitting the regression
results_log = reg_log.fit()

# Creating a logit function, depending on the input and coefficients
def f(x,b0,b1):
    return np.array(np.exp(b0+x*b1) / (1 + np.exp(b0+x*b1)))

# Sorting the y and x, so we can plot the curve
f_sorted = np.sort(f(x1,results_log.params[0],results_log.params[1]))
x_sorted = np.sort(np.array(x1))
ax = plt.scatter(x1,y,color='C0')
#plt.xlabel('SAT', fontsize = 20)
#plt.ylabel('Admitted', fontsize = 20)
# Plotting the curve
ax2 = plt.plot(x_sorted,f_sorted,color='red')
plt.figure(figsize=(20,20))
plt.show()
```
