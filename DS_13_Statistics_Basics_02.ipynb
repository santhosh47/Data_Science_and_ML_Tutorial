## **Variance:**

Variance is a measure of how data points differ from the mean. According to Layman, a variance is a measure of how far a set of data (numbers) are spread out from their mean (average) value.

Variance means to find the expected difference of deviation from actual value. Therefore, variance depends on the standard deviation of the given data set. 

The more the value of variance, the data is more scattered from its mean and if the value of variance is low or minimum, then it is less scattered from mean. Therefore, it is called a measure of spread of data from mean.

# **Standard Deviation**

Since Varaince is the square of the diffrecenre of a data point from the mean it is usually quite large. And it cannot be used directly in statistics to obtain any valuable info. So the squared root of variance is calculated and is called as Standard Deviation which is a good measure of distribution of data points from mean.

Correlation adjusts covariance so that the relationship between two variables becomes easy any intutive to interpret.

Mathematically a correaltion coeffecient value can never be greater than 1 or lesser than -1

**Perfect Positive Correlation** : Correlation is equal to 1 , which mean the entire variable can be dfined by the other variable


**Perfect Negative Correlation** : Correlation is equal to -1 , which mean the entire variable can be dfined by the other variable but in opposite direction. That is if one variable increases the other decreases

**Correlation of 0** : This means the variables are independent of each other

### **It is common knowledge to disregard correlation that are less than 0.2**
