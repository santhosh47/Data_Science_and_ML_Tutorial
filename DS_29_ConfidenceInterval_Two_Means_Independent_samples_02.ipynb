This lesson is about **independent samples with variances unknown, but assumed to be equal.**

## **EXAMPLE**

You are trying to calculate the difference
of the price of apples in New York and LA.
You go to 10 grocery shops in New York
and your friend Paul who lives in LA
visits eight grocery shops
in order to help you with the research.

Once you've organized the data in a table,
you start reflecting on how you can create
a confidence interval that shows the difference
between the price of apples in New York and LA.

image.png
You don't know what the population variance
of apples in New York or LA is,
but you assume it should be the same for NY and LA.

So you calculate the mean price in NY and LA
and obtained $3.94 

and $3.25 respectively.
Moreover, their sample standard deviations
are 18 cents and 27 cents.

**What should we do now?**

Well, we assume that the population variances are equal
so we have to estimate them.

The unbiased estimator in this case
is called the **pooled sample variance**
and we could use the following formula to calculate it.

image.png

As you can see, it is based solely on the sample sizes
and the sample standard deviations of the two datasets.

We quickly plug in the numbers
and get a pooled sample variance of 0.05
and a pooled standard deviation of 22 cents.

### **How about the statistic**
needed to form a confidence interval?

Well, we have an unknown variance, so you guessed it.
It's the T statistic.

Here's the formula for the confidence interval.

image.png

Let's compare it to the formula
for independent samples with known variance.
We have the same difference of sample means,
but the variance is instead of the population variances,
we have the pooled sample variance.
And then instead of the Z statistic,
we have the T statistic.

Although the formulas are different,
they are very consistent, right?

You must be wondering about the T statistic though.
It is a bit stranger this time,
so let's quickly clarify it.

The degrees of freedom are equal to the total sample size
minus the number of variables.

So far, we have seen T's with n minus one degrees of freedom
because we had a sample size of n and only one variable.
This time we have two sample sizes,
one of 10 and one of eight observations and two variables,
apple prices in New York and apple prices in LA.

Therefore, we have 16 degrees of freedom.
All right, checking the T statistic from the table
for a 95% confidence interval with 16 degrees of freedom,
we get 2.12.
Let's plug everything in and get our answer.
The 95% confidence interval is between 0.47 and 0.92.

image.png


**What's the interpretation?**

Well, we are 95% confident that the actual difference
between the two populations,
price of apples in New York and in LA,
is somewhere between 0.47 and 0.92.
Therefore, it is clear that apples in New York
are much more expensive than in LA.
